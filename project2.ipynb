```python
data = pd.read_csv("/Users/darren/Downloads/cvproject/dataset/netflix_titles-1 (1).csv")
data
```

```python
data.describe()
```

```python
data["date_added"] = pd.to_datetime(data["date_added"])

data
```

```python
movie = data[data["type"] == "Movie"]
tv = data[data["type"] == "TV Show"]

max_movie = movie.groupby("release_year")["type"].count()
max_tv = tv.groupby("release_year")["type"].count()

print(max_movie.idxmax())
print(max_tv.idxmax())
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1.5,0.9])

axes.bar(max_movie.index, max_movie.values)
axes.set_xlabel('Year', labelpad = 30)
axes.set_ylabel('Count', labelpad = 30)
plt.xticks(max_movie.index, rotation = 70)
plt.title("Movie")
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1.5,0.9])

axes.bar(max_tv.index, max_tv.values)
axes.set_xlabel('Year', labelpad = 30)
axes.set_ylabel('Count', labelpad = 30)
plt.xticks(max_tv.index, rotation = 70)
plt.title("TV Show")
plt.show()
```

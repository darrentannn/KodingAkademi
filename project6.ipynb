```python
# QUESTION 1d

dataset["Product line"]. unique()
```

```python
dataset["Product line"].replace({"Electronic accessories" : 1, 
                                 "Fashion accessories" : 2,
                                 "Food and beverages" : 3,
                                 "Health and beauty" : 4,
                                 "Home and lifestyle" : 5,
                                 "Sports and travel" : 6},
                                   inplace= True)
```

```python
dataset
```

```python
# QUESTION 1e

dataset["Payment"].replace({"Cash" : 1, 
                            "Credit card" : 2,
                            "Ewallet" : 3},
                            inplace= True)
```

```python
dataset
```

```python
# QUESTION 2

dataset["Total"] = dataset["Unit price"] * dataset["Quantity"]
dataset["Tax"] = dataset["Total"] * 0.05
dataset["Final Price"] = dataset["Total"] - dataset["Tax"]
```

```python
dataset
```

```python
# QUESTION 3

dataset.isnull().sum()
```

```python
# QUESTION 4a

dataset.columns
```

```python
produk = dataset.groupby("Product line")["Quantity"].sum()
produk
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x = produk.index, y = produk.values, label = "Quantity", width = 0.4)
plt.legend()
plt.show()
```

```python
# QUESTION 4b

normal = dataset.loc[dataset["Customer type"]== 1]
normal_group = normal.groupby("Product line")["Customer type"].count()
normal_group
```

```python
plt.figure(figsize =(8,5))
sns.barplot(x = normal_group.index, y = normal_group.values, label= "Customer type", width = 0.6)
plt.legend()
plt.show()
```

```python
# QUESTION 4c

member = dataset.loc[dataset["Customer type"] == 2]
member_group = member.groupby("Product line")["Customer type"].count()
member_group
```

```python
plt.figure(figsize =(8,5))
sns.barplot(x = member_group.index, y = member_group.values, label= "Customer type", width = 0.6)
plt.legend()
plt.show()
```

```python
# QUESTION 4C

payment = dataset.groupby("Payment")["Invoice ID"].count()
payment
```

```python
plt.figure(figsize =(8,5))
sns.barplot(x = payment.index, y = payment.values, label= "Invoice ID", width = 0.6)
plt.legend()
plt.show()
```

```python
# QUESTION 4d

plt.pie(payment.values, labels = payment.index, autopct='%1.1f%%')
plt.title('Payment')
plt.axis('equal')
plt.legend()
plt.show()
```

```python
# QUESTION 4e

profit = dataset.groupby("Product line")["Final Price"].sum()

profit
```

```python
plt.figure(figsize =(8,5))
sns.barplot(x = profit.index, y = profit.values, label= "Final Price", width = 0.6)
plt.legend()
plt.show()
```
